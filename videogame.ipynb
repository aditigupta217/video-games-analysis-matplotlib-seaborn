```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('vgsales.csv')
data.head()
```

```python
# 1. univariate analysis
# categorical
```

```python
data['Genre'].unique()
```

```python
cat_count=data['Genre'].value_counts()
```

```python
cat_count
```

```python
X_bar = cat_count.index
Y_bar = cat_count.values
```

```python
plt.figure(figsize=(12,8))
plt.bar(X_bar,Y_bar)
```

```python
plt.xticks(rotation=90,fontsize = 10)
plt.bar(X_bar,Y_bar,width = 0.5, color='Red')
plt.title('top video games')
plt.show()
```

```python
sns.countplot(x='Genre',data=data,order=data['Genre'].value_counts().index)
plt.xticks(rotation=90,fontsize = 10)
plt.show()
```

```python
sales_data=data[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales' ]]
```

```python
sales_data
```

```python
region_sales=sales_data.T.sum(axis='columns')
```

```python
region_sales
```

```python
plt.pie(region_sales, labels=region_sales.index, startangle=90, explode=(0.2,0,0,0) ,autopct='%1.1f%%')
plt.title('Total Sales across various regions')
plt.show()
```

```python
# categorical data 
# top 5 publisher
top_pub = data['Publisher'].value_counts().index[:5]
top_gen = data['Genre'].value_counts().index[:5]
top_plat = data['Platform'].value_counts().index[:5]
top_plat
```

```python
top_pub
```

```python

top_gen
```

```python
top_5_data = data.loc[(data['Publisher'].isin(top_pub))&(data['Platform'].isin(top_plat))&(data['Genre'].isin(top_gen))]
```

```python
top_5_data
```

```python
#dodgd count plot
sns.countplot(data=top_5_data,x='Publisher', hue='Platform')
plt.xticks(rotation=90,fontsize = 10)
```

```python
# converted to pivot table
df_stacked_plot= pd.crosstab(index=top_5_data['Publisher'],columns=top_5_data['Platform'])
df_stacked_plot
```

```python
df_stacked_plot.plot(kind='bar', stacked=True)
```

```python
sns.boxplot(x='Publisher',y='Global_Sales', data=top_5_data)
plt.xticks(rotation=90,fontsize = 10)
```
